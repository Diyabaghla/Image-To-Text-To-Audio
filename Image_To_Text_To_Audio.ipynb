```python
from PIL import Image
import pytesseract
import gtts
```

```python
img=Image.open("teddy1.jpg")
print(img)
```

```python
result=pytesseract.image_to_string(img)
print(result)
```

```python
audio=gtts.gTTS(result)
audio.save("teddy.mp3")
```
